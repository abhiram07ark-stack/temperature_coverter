```python
print("Temperature Converter")
choice=int(input(
    "1.Convert from Farhneit to Celsius:\n"
    "2.Celsius to Farhneit:\n"
    "3.Kelvin to Celsius:\n"
    "4.Celsius to Kelvin:\n"
"Choose from 1 - 4\n"))
value=float(input("Enter the value to be converted"))
if choice==1:
    print("Convert value from Farhneit to Celsius")
    print((value-32)*5/9)
elif choice==2:
    print("Convert Celsius to Farhneit")
    print((value*9/5)+32)
elif choice==3:
    print("Convert Kelvin to Celsius")
    print(value-273.15)
elif choice==4:
    print("Convert Celsius to Kelvin")
    print(value+273.15)
else:
    print("Invalid Number")
```
